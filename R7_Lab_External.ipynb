### Build a DNN using Keras with `RELU` and `ADAM`

#### Load tensorflow

```
%tensorflow_version 2.x
import tensorflow as tf
```

#### Collect Fashion mnist data from tf.keras.datasets 

```
from keras.datasets import fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

#### Change train and test labels into one-hot vectors

```
train_y = tf.keras.utils.to_categorical(y_train, dtype='float32')
test_y = tf.keras.utils.to_categorical(y_test, dtype='float32')
```

#### Build the Graph

#### Initialize model, reshape & normalize data

```
#normalize data
x_train_norm = x_train/255
x_test_norm = x_test/255
```

```
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Reshape
```

```
#Clear out tensorflow memory
tf.keras.backend.clear_session()

#Initialize Sequential model
model = Sequential()

#Reshape data from 2D to 1D -> 28x28 to 784
model.add(Reshape((784,),input_shape=(28,28,)))

#Normalize the data
model.add(tf.keras.layers.BatchNormalization())
```

#### Add two fully connected layers with 200 and 100 neurons respectively with `relu` activations. Add a dropout layer with `p=0.25`

```
#Add 1st hidden layer
model.add(Dense(200, activation='relu'))

#Add 2nd hidden layer
model.add(Dense(100, activation='relu'))

#Add another dropout layer
model.add(Dropout(0.25))
```

### Add the output layer with a fully connected layer with 10 neurons with `softmax` activation. Use `categorical_crossentropy` loss and `adam` optimizer and train the network. And, report the final validation.

```
#Add OUTPUT layer
model.add(Dense(10, activation='softmax'))
```

```
#Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```
#Train the model
model.fit(x_train_norm,train_y,validation_data=(x_test_norm,test_y),epochs=10,batch_size=32)
```

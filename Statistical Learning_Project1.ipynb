```python
import pandas as pd
import numpy as np
df=pd.read_csv('Laliga.csv')
df.columns = df.iloc[0]
df = df.reindex(df.index.drop(0)).reset_index(drop=True)
df.columns.name = None
df.replace(to_replace='-',value=0, inplace=True)
cols = ['Pos', 'Seasons', 'Points', 'GamesPlayed', 'GamesWon',
        'GamesDrawn', 'GamesLost', 'GoalsFor', 'GoalsAgainst', 'Champion',
        'Runner-up', 'Third', 'Fourth', 'Fifth', 'Sixth', 'T', 'BestPosition']
df[cols] = df[cols].apply(pd.to_numeric, axis=1)
df
```

```python
df_cond = df[(df['Debut']>'1930-31') & (df['Debut']<'1979-80')]
df_cond
```

```python
df.sort_values(by=['Team','Points'],ascending=False).head(5)
df.iloc[0:5][['Team', 'Points']]
```

```python
def Goal_diff_count(F1,F2):
    return F1-F2
```

```python
df['Goal_diff_count']=Goal_diff_count(df['GoalsFor'],df['GoalsAgainst'])
print('The team which has highest goal difference is :', df.iloc[df['Goal_diff_count'].idxmax(),1])
```

```python
df['WinningPercent']=df['GamesWon']/df['GamesPlayed']*100
df['WinningPercent'].fillna(value='0', inplace=True )
df
```

```python
df['WinningPercent']=df['WinningPercent'].apply(pd.to_numeric)
df.sort_values(by=['Team', 'WinningPercent'],ascending=False)
df.iloc[0:5][['Team', 'WinningPercent']]
```

```python
Best_position=df.groupby("Team")["Pos", "Points"].agg("sum").sort_values("Points", ascending=False)
Best_position
```

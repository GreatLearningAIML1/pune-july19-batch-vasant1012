### A MNIST-like fashion product database

In this, we classify the images into respective classes given in the dataset. We use a Neural Net and a Deep Neural Net in Keras to solve this and check the accuracy scores.

### Load tensorflow

```python
import tensorflow as tf
tf.set_random_seed(42)
```

```python
tf.__version__
```

### Collect Data

```python
import keras
```

```python
(trainX, trainY), (testX, testY) = keras.datasets.fashion_mnist.load_data()
```

```python
print(testY[0:5])
```

### Convert both training and testing labels into one-hot vectors.

**Hint:** check **tf.keras.utils.to_categorical()**

```python
trainY = tf.keras.utils.to_categorical(trainY, num_classes=None, dtype='float32')
testY = tf.keras.utils.to_categorical(testY, num_classes=None, dtype='float32')
```

```python
print(trainY.shape)
print('First 5 examples now are: ', trainY[0:5])
```

### Visualize the data

Plot first 10 images in the triaining set and their labels.

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize = (8,8))
for i in range(1,11):
  fig.add_subplot(1,10,i)
  plt.imshow(trainX[i])
plt.show()
print("label for each of the above image:")
print(trainY[:11])
```

### Build a neural Network with a cross entropy loss function and sgd optimizer in Keras. The output layer with 10 neurons as we have 10 classes.

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
model.compile(optimizer='SGD', loss='categorical_crossentropy',metrics=['accuracy'])
```

### Execute the model using model.fit()

```python
model.fit(trainX, trainY, epochs=10)
```

### In the above Neural Network model add Batch Normalization layer after the input layer and repeat the steps.

```python
model_1 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax'),
    keras.layers.BatchNormalization()
])
```

### Execute the model

```python
model_1.compile(optimizer='SGD', loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model_1.fit(trainX, trainY, epochs=10)
```

### Customize the learning rate to 0.001 in sgd optimizer and run the model

```python
model_2 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax'),
    keras.layers.BatchNormalization()
])
```

```python
from keras import optimizers
sgd = optimizers.SGD(lr=0.001)
```

```python
model_2.compile(optimizer=sgd, loss='categorical_crossentropy', metrics=['accuracy'])
model_2.fit(trainX, trainY, epochs=10)
```

### Build the Neural Network model with 3 Dense layers with 100,100,10 neurons respectively in each layer. Use cross entropy loss function and singmoid as activation in the hidden layers and softmax as activation function in the output layer. Use sgd optimizer with learning rate 0.03.

```python
model_final = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='sigmoid'),
    keras.layers.Dense(100, activation='sigmoid'),
    keras.layers.Dense(10, activation='softmax'),
    keras.layers.BatchNormalization()
])
```

```python
sgd_final = optimizers.SGD(lr=0.03)
```

```python
model_final.compile(optimizer=sgd_final, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model_final.fit(trainX, trainY, epochs=10)
```

## Review model

```python
model_final.summary()
```

### Run the model

```python
model_final.fit(trainX, trainY, validation_data=(testX, testY), epochs=30, batch_size = 32)
```
